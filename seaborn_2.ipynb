```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

## Categorical plots <br>
Categorical scatter plot (bivariate analysis) <br>1.Stripplot<br> 2.SwarmPlot<br>
Categorical Distribution plot (monovariate analysis) <br>1.Box plot<br> 2.Violoin plot<br>
Categorical estimation plot (to calculate aggregate values)<br> 1.Barplot<br> 2.Point plot<br> 3.Count plot

```python
#strip plot
tips
```

```python
#axeslevel
sns.stripplot(data=tips,x='day',y='total_bill',hue= 'day',jitter=False)          #jitter for noise
 
```

```python
#figure level
sns.catplot(data=tips,x='day',y='total_bill',hue= 'day',kind = 'strip',jitter=False)  
```

```python
#figure level
sns.catplot(data=tips,x='day',y='total_bill',hue= 'day',kind = 'strip',jitter=0.2)  
```

```python
#swarmplot
#figure level
sns.catplot(data=tips,x='day',y='total_bill',hue= 'day',kind = 'swarm')  
```

```python
#axes level
sns.swarmplot(data=tips,x='day',y='total_bill',hue= 'day')  #This uses an algo which gives better idea of dist. ,good for small datasets
```

## box plot 
A boxplot is a standardized way of displaying the distribution of data based on a five number 
summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”).
It can tell you about your outliers and what their values are. Boxplots can also tell 
you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.


```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box',hue='sex')
```

```python
sns.catplot(data=tips,y='total_bill',kind='box')
```

## Violin plot (Box plot+KDE plot)

```python
#Bar plot
sns.barplot(data=tips,x='day',y='total_bill')
```

```python
#Bar plot
sns.barplot(data=tips,x='day',y='total_bill',ci=None)
```

```python
#Bar plot
import numpy as np
sns.barplot(data=tips,x='day',y='total_bill',ci=None,estimator=np.std)   #by default its mean.
```

```python
#pointplot
#Bar plot
import numpy as np
sns.pointplot(data=tips,x='day',y='total_bill',errorbar=None,estimator=np.min)   #by default its mean.
```

```python
#countplot--like histogram and valuecount in pandas
sns.countplot(data=tips,x='sex')
```

```python
#faceting using catplot
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',kind='box')
```

```python
#faceting using catplot
sns.catplot(data=tips,x='sex',y='total_bill',col='smoker',kind='box',row='time')
```

## Regression Plots<br> 
regplo <br>t l
lmpl<br>ot
In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
tips
```

```python
#axes level
#hue parametre isnt available
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
#resid plot
#the pts are the vertical distances from the regression line.i.e, error in prediction
sns.residplot(data=tips,x='total_bill',y='tip')
```
