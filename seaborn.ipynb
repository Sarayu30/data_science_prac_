```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
!pip install plotly
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#axes level 
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
#figure level-- recommeneded!
sns.relplot(data=tips,x='total_bill',y='tip',kind = 'scatter')
```

```python
gap = px.data.gapminder()
gap
temp_df=gap[gap['country']=='India']
```

```python
temp_df
```

```python
#axes level
sns.lineplot(data=temp_df,x='year',y='lifeExp' )
```

```python
#figure level
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','China','Pakistan'])]
```

```python
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
temp_df = gap[gap['country'].isin(['India','China','Germany'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country',style='continent')
```

```python
#facet plot
 
sns.relplot(data=tips,x='total_bill',y='tip',kind = 'scatter',col='sex')
```

```python
#facet plot
 
sns.relplot(data=tips,x='total_bill',y='tip',kind = 'scatter',row='sex')
```

```python
#facet plot --only work with figure level-relplot
 
sns.relplot(data=tips,x='total_bill',y='tip',kind = 'scatter',col='sex',row='smoker')
```

```python
# figure level : displot <br>
# axes level : histplot,kdeplot,rigplot
```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.histplot(data=tips,x='total_bill',bins=20)
```

```python
sns.histplot(data=tips,x='day')  # can be used for catgorical data also,unlike in other lib.This is actually a count plot
```

```python
sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic= sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic,x='age',kind='hist',hue='sex',element='step')
```

```python
#kernel density estimation :Rather than using discrete bins,a KDE PLOT smoothens the obs with a gaussian kernel,producing cont. density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.kdeplot(data=tips,x='total_bill',hue='sex',fill=True)
```

```python
## rug plot-plot marginal dist,by drawing ticks along x,y axes  
## Just compliments other plots
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
## bivariate histogram
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
sns.kdeplot(data=tips,x='total_bill',y='tip')
```

## Matrix Plot


```python
## Heat Map plots rectangular data as color coded matrix
```

```python
## convert into wide data format
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
 
```

```python
gap[gap['continent']=='Europe']
```

```python
temp_df=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,cmap='summer')
 
```
