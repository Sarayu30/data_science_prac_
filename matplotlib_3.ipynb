## BarChart
1.Bivariate Analysis 2.Numerical vs Ctegorical 3. usecase- Aggregate analysis of groups

```python
df=pd.read_csv("C:\\Users\\sarayu krishna\\Downloads\\batsman_season_record.csv")
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
plt.bar(np.arange(df.shape[0]) -0.2,df['2015'],width = 0.2)
plt.bar(np.arange(df.shape[0]) ,df['2016'],width = 0.2)
plt.bar(np.arange(df.shape[0]) +0.2,df['2017'],width = 0.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(np.arange(df.shape[0]), df['2015'], width = 0.2)
axs[1].bar(np.arange(df.shape[0]), df['2016'], width = 0.2)
axs[2].bar(np.arange(df.shape[0]), df['2017'], width = 0.2)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 3))

# Data
bar_width = 0.2
bar_positions_2015 = np.arange(df.shape[0])
bar_positions_2016 = bar_positions_2015 + bar_width
bar_positions_2017 = bar_positions_2016 + bar_width

# Plotting
ax.bar(bar_positions_2015, df['2015'], width = bar_width, label='2015')
ax.bar(bar_positions_2016, df['2016'], width = bar_width, label='2016')
ax.bar(bar_positions_2017, df['2017'], width = bar_width, label='2017')

# xticks
ax.set_xticks(bar_positions_2016)
ax.set_xticklabels(df.index)

# Labels and title
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by group and year')

# Legend
plt.legend()

plt.show()
```
